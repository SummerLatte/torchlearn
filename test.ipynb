```python
# y = XW + b
# y = 2X + 3

import numpy as np
import matplotlib.pyplot as plt
import torch
from d2l import torch as d2l

def lineg(W, b):
    X = torch.normal(0, 10, (1000, len(W)))
    y = torch.matmul(X, W) + b
    y = y + torch.normal(0, 3, y.shape)
    return X, y

```

```python
feature, label = lineg(torch.tensor([2.0]), 3.0)

d2l.set_figsize()
d2l.plt.scatter(feature[:].numpy(), label.numpy(), 1)
```

```python
import random

def data_itr(features, label, batchsize):
    number = len(features)
    indeces = list(range(number))
    random.shuffle(indeces)
    for i in range(0, number, batchsize):
        batch = torch.tensor(indeces[i:min(i+batchsize, number)])
        yield features[batch], label[batch]
```

```python
data_itr(feature, label, 10)
for X, y in data_itr(feature, label, 10):
    print(X, y)
    break
```
